```python
import re
import nltk
from datasets import load_dataset
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

# Download necessary NLTK data
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Text cleaning function (without removing special characters)
def clean_text(text):
    text = text.lower()  # Lowercase
    text = re.sub(r'\s+', ' ', text).strip()  # Remove extra whitespace
    return text

# Preprocess text function (without removing special characters)
def preprocess_text(text):
    # Clean text
    text = clean_text(text)

    # Tokenize text
    tokens = word_tokenize(text)

    # Remove stop words
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]

    # Lemmatize tokens
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token) for token in tokens]

    # Join tokens back to string
    return ' '.join(tokens)

# Preprocess the dataset
def preprocess_dataset(dataset):
    dataset = dataset.map(lambda x: {'article': preprocess_text(x['article'])})
    return dataset

# Specify the cache directory on your D: drive
cache_dir = "D:/datasets_cache"

# Load the PubMed Summarization dataset and specify cache_dir
ds = load_dataset("ccdv/pubmed-summarization", "document", cache_dir=cache_dir)

# Preprocess the training, validation, and test sets
ds['train'] = preprocess_dataset(ds['train'])
ds['validation'] = preprocess_dataset(ds['validation'])
ds['test'] = preprocess_dataset(ds['test'])

# Check the preprocessed dataset
print(ds['train'][0])
```
